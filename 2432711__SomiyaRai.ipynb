**Task - 3 - Outlier Detection:**

**Tasks:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

2. Define outliers using the 1.5 Ã— IQR rule.

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.


```python
import seaborn as sns
```

```python
# Task 1&2: Calculate IQR for Score and GDP per Capita and define outlier using 1.5 X IQR rule

Q1_score = dataset['score'].quantile(0.25)
Q3_score = dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers_score = dataset[(dataset['score'] < lower_bound_score) | (dataset['score'] > upper_bound_score)]
outliers_gdp = dataset[(dataset['Log GDP per capita'] < lower_bound_gdp) | (dataset['Log GDP per capita'] > upper_bound_gdp)]

outliers_combined = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

print(f"Outliers based on Score:\n{outliers_score[['Country name', 'score']]}")
print(f"Outliers based on GDP per Capita:\n{outliers_gdp[['Country name', 'Log GDP per capita']]}")
```

```python
# Task 3: Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='Log GDP per capita', y='score', label='Normal', color='blue')
sns.scatterplot(data=outliers_combined, x='Log GDP per capita', y='score', label='Outliers', color='red')
plt.title('Scatter Plot of GDP per Capita vs. Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

**6. Outlier Detection:**

â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.

â€¢ Plot these outliers and discuss their implications.


```python
# 6. Outlier Detection.

def identify_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

metrics = ['score', 'Log GDP per capita']

print("Outliers in South Asia Region:")
for metric in metrics:
    outliers_sa = identify_outliers(south_asia_data, metric)
    print(f"\nOutliers based on {metric}:")
    print(outliers_sa[['Country name', metric]])

    # Scatter plot for South Asia Region
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=south_asia_data[metric], y=south_asia_data['score'], color='blue', label='Data Points', alpha=0.7)
    sns.scatterplot(x=outliers_sa[metric], y=outliers_sa['score'], color='red', label='Outliers', s=100, marker='X')
    plt.title(f"South Asia: {metric} vs Score with Outliers")
    plt.xlabel(metric)
    plt.ylabel("Score")
    plt.legend()
    plt.grid(True)
    plt.show()

print("\nOutliers in Middle East:")
for metric in metrics:
    outliers_me = identify_outliers(middle_east, metric)
    print(f"\nOutliers based on {metric}:")
    print(outliers_me[['Country name', metric]])

    # Scatter plot for Middle East Region
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=middle_east[metric], y=middle_east['score'], color='green', label='Data Points', alpha=0.7)
    sns.scatterplot(x=outliers_me[metric], y=outliers_me['score'], color='red', label='Outliers', s=100, marker='X')
    plt.title(f"Middle East: {metric} vs Score with Outliers")
    plt.xlabel(metric)
    plt.ylabel("Score")
    plt.legend()
    plt.grid(True)
    plt.show()
```

3. Outliers: The South Asia has an outlier below the score of 2, indicating a particularly low score that stands out from the rest of the data. Meanwhile, Middle East has no outlier, suggesting a more uniform spread of scores.
